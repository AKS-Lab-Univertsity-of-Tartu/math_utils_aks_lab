```python
import jax
from lp_butterworth_filter import lowpass_filter
```

```python
T = 100
key = jax.random.PRNGKey(0)

eta = jax.random.normal(key, shape=(T,))
eta_filtered = lowpass_filter(eta, dt=0.1, fc=1.0)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(eta)
plt.plot(eta_filtered)
plt.show()
```
